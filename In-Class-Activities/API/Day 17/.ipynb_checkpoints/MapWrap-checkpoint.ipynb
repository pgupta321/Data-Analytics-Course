```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import openweathermapy as ow
import pandas as pd
from pprint import pprint
```

```python
# Create a settings object with your API key and preferred units

settings = {'units' : 'imperial', 'appid' : 'fc13d9f6ab3c9f2edef50aea529070a2'}
```

```python
# Get data for each city in cities.csv

cf = pd.read_csv('cities.csv', header=None)

# Getting data from the first and only column

cities = [city for city in cf[0]]

cf
```

```python
weather_data = []

#weather_data.append(ow.get_current(city, **settings) for city in cities)

weather_data = [ow.get_current(city, **settings) for city in cities]

weather_data
```

```python
# Create an "extracts" object to get the temperature, latitude,
# and longitude in each city

extracts = ["main.temp", "coord.lat", "coord.lon"]

extract_data = [city(*extracts) for city in weather_data]


# Create a Pandas DataFrame with the results

df = pd.DataFrame(extract_data, index = cities, columns = ['Temperature', 'Latitude', 
                                                          'Longitude'])
df
```
