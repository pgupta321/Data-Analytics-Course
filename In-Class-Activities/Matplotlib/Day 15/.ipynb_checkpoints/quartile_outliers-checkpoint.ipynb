```python
# Dependencies
import matplotlib.pyplot as plt
from stats import median
import numpy as np
```

```python
### Data Points
arr = np.array([2.3, 10.2,11.2, 12.3, 14.5, 14.6, 15.0, 15.1, 19.0, 24.0])
arr
```

```python
# Find the median
mid = median(arr)
mid
```

```python
# Use numpy to create quartiles
q1 = np.percentile(arr, 25)
q3 = np.percentile(arr, 75)
```

```python
# Print the quartiles
print(f'Q1 is {q1}, Q3 is {q3}')
```

```python
# Calculate the interquartile range
iqr = q3 - q1

print(f'IQR: {iqr}')
```

```python
# Find lower boundary
# Q1 - 1.5 * IQR

lower_boundary = q1 - 1.5 * iqr
lower_boundary
```

```python
# Find upper boundary
# Q3 + 1.5 * IQR

upper_boundary = q3 + 1.5 * iqr
upper_boundary
```

```python
# Check for any lower outliers

arr[arr<=lower_boundary]
```

```python
# Check for any upper outliers

arr[arr>=upper_boundary]
```

```python
# Create box plot

plt.boxplot(arr, showmeans=True)
plt.grid()
plt.show()
```
