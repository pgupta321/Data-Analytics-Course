```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
path = "avg_rain_state.csv"
df = pd.read_csv(path)
```

```python
# Set x axis and tick locations
states = df['State']
inches = df["Inches"]

x_axis = np.arange(len(df))
tick_locations = [x+0.4 for x in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize = (20, 3))
plt.bar(x_axis, inches, color = 'red', alpha = 0.5, align = 'edge')
plt.xticks(tick_locations, states, rotation = 'vertical')
```

```python
# Set x and y limits

plt.xlim(-1.25, len(x_axis))
plt.ylim(0, max(inches) + 10)
```

```python
# Set a Title and labels

plt.title('Average Rain per State')
plt.xlabel('State')
plt.ylabel('Average Amount of Rainfall (in)')
```
