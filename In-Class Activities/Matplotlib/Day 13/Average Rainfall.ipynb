```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
path = "avg_rain_state.csv"
df = pd.read_csv(path)
```

```python
# Set x axis and tick locations
states = df['State']
inches = df["Inches"]

x_axis = np.arange(len(df))
tick_locations = [x+0.4 for x in x_axis]

plt.xticks(tick_locations, states)

plt.bar(x_axis, inches)
```
