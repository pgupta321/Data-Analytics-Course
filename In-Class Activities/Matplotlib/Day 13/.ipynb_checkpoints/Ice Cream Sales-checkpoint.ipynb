```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
```

```python
scoop_price = [100, 150, 50, 140, 200, 400, 300, 600, 500, 200, 400, 200]
facecolors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'red', 'blue', 'green', 'orange', 'purple', 'pink']
```

```python
# Tell matplotlib to create a scatter plot based upon the above data
plt.scatter(temp, sales, marker="o", facecolors=facecolors, edgecolors="black", s = scoop_price)
```

```python
# Set the upper and lower limits of our y axis
```

```python
# Set the upper and lower limits of our x axis
plt.xlim(11, 26)
```

```python
# Create a title, x label, and y label for our chart
plt.title("Ice Cream Sales v Temperature")
plt.xlabel("Temperature (Celsius)")
plt.ylabel('Sales (Dollars)')
```
