```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 130, 10)
DangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.plot(time, DangerDrop, color = 'red', label = 'Danger Drop')
plt.plot(time, RailGun, color = 'green', label = 'Rail Gun')
```

```python
plt.xlim(0, 120)
plt.ylim(5, 95)
```

```python
plt.legend(loc = 'best')
```

```python
plt.xlabel('Coaster Runtime')
plt.ylabel('Speed (MPH)')
```

```python
plt.grid()
```

```python
plt.title('Coaster Speed Over Time')
```
