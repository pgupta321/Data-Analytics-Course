```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(1, 13, 1)
```

```python
y_axis = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(x_axis, y_axis)
```

```python
celsius_data = [(y - 32) * 0.56 for y in y_axis]
```

```python
plt.plot(x_axis, celsius_data)
```

```python
plt.plot(x_axis, y_axis)
plt.plot(x_axis, celsius_data)
```

```python
F_handle = plt.plot(x_axis, y_axis, marker = '+', color = 'blue', label = 'Fahrenheit')
C_handle = plt.plot(x_axis, celsius_data, marker = 's', color = 'red', label = 'Celsiuis')
```

```python
plt.xlabel('Months')
plt.ylabel('Degrees')
```

```python
plt.legend(loc = 'lower right')
```
