```python
# Dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV

path = "05-Matplotlib_Day14_03-Stu_BattlingKings_Unsolved_Resources_got.csv"

file = pd.read_csv(path)

file
```

```python
# Get attacker and defender data

attacker = file['attacker_king'].value_counts()
attacker.head()

defender = file['defender_king'].value_counts()
print(defender)
```

```python
# Get total battle data
totalbattle = attacker.add(defender, fill_value = 0)
```

```python
# Configure plot and ticks

totalbattle.plot(kind = 'bar', color = 'red')
```

```python
# Set textual properties

plt.xlabel('Kings')
plt.ylabel('Number of Battles')
plt.title("The Bloodthirst of Kings")
```
