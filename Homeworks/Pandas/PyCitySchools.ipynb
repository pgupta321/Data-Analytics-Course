# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with the two highest ranges of spending per student actually underperformed compared to schools with smaller budgets.

* Smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%). Overall passing rates dropped 20% (94% to 74%) in the large schools, compared to small and medium-sized schools.

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 

* The average math and reading scores across all 4 grades are largely similar within each school, with a slightly wider range of scores in math than in reading.

* In the worst performing schools, there is a significant difference (11%) between students passing math and students passing reading, with reading having the higher percentage. This discrepancy is smaller in the best performing schools, with students passing reading being 3% more than students passing math.

* In the district as a whole, there are 11% more students passing reading than students passing math, even though the average reading score is only 3 points higher than the average math score. This could be due to the fact that the average math score is only 9 points away from a passing grade.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import statistics as st

# File to Load
school_data_to_load = "04-Pandas_Homework_PyCitySchools_Resources_schools_complete.csv"
student_data_to_load = "04-Pandas_Homework_PyCitySchools_Resources_students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate total number of schools

totalschools = len(school_data["school_name"].unique())

# Calculate total number of students

totalstudents = len(school_data_complete["Student ID"].unique())

# Calculate total budget and convert it to currency format

totalbudget = school_data["budget"].sum()
tbformat = ('$' + format(totalbudget, ',.2f'))

#totalbudget = totalbudget.map("${:,.2f}".format) - not working

print(tbformat)
```

```python
# Calculate average math score

avgmath = round(school_data_complete["math_score"].mean(), 6)

# Calculate average reading score

avgread = round(school_data_complete["reading_score"].mean(), 6)
```

```python
# Calculate number of students passing math, percentage of students passing math, and convert to percent format

passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]

passing_math_rate = round(passing_math_count / float(totalstudents) * 100, 6)

passing_math_percent = str(passing_math_rate) + "%"
```

```python
# Same process with students passing reading

passing_read_count = school_data_complete[(school_data_complete['reading_score'] >= 70)].count()['student_name']

passing_read_rate = round(passing_read_count / float(totalstudents) * 100, 6)

passing_read_percent = str(passing_read_rate) + "%"
```

```python
# Calculate passing rates for math and reading, overall passing percent, and convert to percent format

passrates = [passing_math_rate, passing_read_rate]

overallpassingrate = round(st.mean(passrates), 6)

overallpassingpercent = str(overallpassingrate) + '%'
```

```python
district_summary_df = pd.DataFrame([{'Total Schools' : totalschools,
                                    'Total Students' : totalstudents,
                                    'Total Budget' : tbformat,
                                    'Average Math Score' : avgmath,
                                    'Average Reading Score' : avgread,
                                    '% Passing Math' : passing_math_percent,
                                    '% Passing Reading' : passing_read_percent,
                                    'Overall Passing Rate' : overallpassingpercent}],
                                   columns=['Total Schools', 'Total Students', 'Total Budget', 
                                              'Average Math Score', 'Average Reading Score',
                                              '% Passing Math', '% Passing Reading', 'Overall Passing Rate'])

district_summary_df.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Make series for School Type and Total Students, using School Name as index

school_types = school_data.set_index(["school_name"])["type"]

schoolsize = school_data_complete["school_name"].value_counts()

print(school_types)
print(schoolsize)
```

```python
# Make list for total school budgets

school_budget = school_data_complete.groupby(['school_name']).mean()['budget']

#school_budget = school_data.set_index(["school_name"])["budget"].sort_index() - another correct method

print(school_budget)
```

```python
# Calculate per student budget

psbudget = school_budget/schoolsize

print(psbudget)
```

```python
# Group by school and calculate means of math and reading scores for each school

school_groups = school_data_complete.groupby(['school_name'])

mathscores = school_groups.mean()['math_score']

readscores = school_groups.mean()['reading_score']

print(mathscores)
print(readscores)
```

```python
# Filter data to get passing math scores, then group by school to get the count of students passing math in each school


#school_math_filter = school_data_complete.loc[school_data_complete['math_score'] >= 70, :] - another method


school_math_filter = school_data_complete[school_data_complete['math_score'] >= 70]

school_groups2 = school_math_filter.groupby(['school_name'])

school_passing_math_count = school_groups2.count()["student_name"]


#school_passing_math_count = school_groups2[(school_groups["math_score"] >= 70)].count()["student_name"] - does not work


print(school_passing_math_count)
```

```python
# Same process to get count of students passing reading in each school

school_read_filter = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]

school_groups3 = school_read_filter.groupby(['school_name'])

school_passing_read_count = school_groups3.count()["student_name"]

print(school_passing_read_count)
```

```python
# Calculate passing math rates for schools

school_passing_math_rate = round(school_passing_math_count / schoolsize * 100, 6)

print(school_passing_math_rate)
```

```python
# Calculate passing reading rates for schools

school_passing_read_rate = round(school_passing_read_count / schoolsize * 100, 6)

print(school_passing_read_rate)
```

```python
# Calculate overall passing rates at each school

schoolpassingrate = (school_passing_math_rate + school_passing_read_rate) / 2

print(schoolpassingrate)
```

```python
# Convert formats to currency for budget variables

school_budget = school_budget.map("${:,.2f}".format)
psbudget = psbudget.map("${:,.2f}".format)
```

```python
school_summary_df = pd.DataFrame({'School Type': school_types,
                                 'Total Students' : schoolsize,
                                 'Total School Budget': school_budget,
                                 'Per Student Budget' : psbudget,
                                 'Average Math Score' : mathscores,
                                 'Average Reading Score' : readscores,
                                 '% Passing Math' : school_passing_math_rate,
                                 '% Passing Reading' : school_passing_read_rate,
                                 '% Overall Passing Rate' : schoolpassingrate})

print(school_summary_df)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins and labels

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Remove percent formatting from "Per Student Budget" columns and convert to float in order to use cut method

school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].str.replace(',', '')
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].str.replace('$', '')
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].astype(float)
```

```python
# Duplicate data frame - need to use copy function, otherwise original df is also affected when slicing
# Perform cut and drop unnecessary columns

spending_range_df = school_summary_df.copy()
spending_range_df["Spending Range"] = pd.cut(spending_range_df["Per Student Budget"], spending_bins, labels=group_names)
spending_range_df.drop(['School Type', 'Total Students', 'Total School Budget', 'Per Student Budget'], axis=1, inplace=True)

spending_range_df
```
